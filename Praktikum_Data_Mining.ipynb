```python
import pandas as pd
```

```python
df = pd.read_excel("dataset.xls",)
```

```python
df = df.iloc[0:5001]
```

```python
original_features = df.columns
print('OriginalFeaturesn Count', len(original_features))
print('OriginalFeatures', original_features)
```

```python
# prompt: seleksi fitur tertentu saja yang digunakan seperti fitur yang digunakan adalah Tanggal Submit dan Institusi saja

new_var = ['No', 'Diagnosa Penyakit', 'Institusi', 'Area Kota', 'Rumah Sakit / Poliklinik', 'Tipe Dokter', 'Total Biaya Penggantian (Rp)']

df = df[new_var]
df.head()
```

```python
# Ganti fitur (Seleksi Fitur)

new_columns = [col.replace('No', 'no').replace('Diagnosa Penyakit', 'diagnosa').replace('Institusi', 'institusi').replace('Area Kota', 'area')
.replace('Rumah Sakit / Poliklinik', 'rumah').replace('Tipe Dokter', 'tipe').replace('Total Biaya Penggantian', 'biaya_berobat')for col in df.columns]
df.columns = new_columns
df.head()
```

```python
# Ubah semua data menjadi lowercase (casefolding)

for column in df.columns:
  if df[column].dtype == object:
    df[column] = df[column].str.lower()

df.head()
```

```python
# Ngecek fitur apa saja yang memiliki data mising

missing_values = df.isnull().sum()
features_with_missing_data = missing_values[missing_values > 0]
print("Fitur dengan data missing:")
print(features_with_missing_data)
```

```python
# prompt: Isi data missing pada rumah dengan menggunakan hubungan dari institusi dan area

# Mengisi data missing pada 'rumah' dengan menggunakan hubungan dari 'institusi' dan 'area'
df['rumah'].fillna(df.groupby(['institusi', 'area'])['rumah'].transform('first'), inplace=True)

# Jika masih ada data missing pada 'rumah' setelah pengisian berdasarkan 'institusi' dan 'area',
# bisa diisi dengan modus dari kolom 'rumah'
df['rumah'].fillna(df['rumah'].mode()[0], inplace=True)

# Cek kembali fitur dengan data missing
missing_values = df.isnull().sum()
features_with_missing_data = missing_values[missing_values > 0]
print("Fitur dengan data missing:")
print(features_with_missing_data)
```

```python
# Isi data missing untuk tipe

# Mengisi data missing pada kolom 'tipe' dengan modus (nilai yang paling sering muncul)
modus_tipe = df['tipe'].mode()[0]
df['tipe'].fillna(modus_tipe, inplace=True)

# Memeriksa kembali apakah masih ada data missing pada kolom 'tipe'
missing_values = df['tipe'].isnull().sum()
print("Jumlah data missing pada kolom 'tipe':", missing_values)
```

```python
# Isi data missing untuk biaya berobat

# Mengisi data missing pada kolom 'biaya_berobat' dengan median (nilai tengah)
median_biaya_berobat = df['biaya_berobat (Rp)'].median()
df['biaya_berobat (Rp)'].fillna(median_biaya_berobat, inplace=True)

# Memeriksa kembali apakah masih ada data missing pada kolom 'biaya_berobat'
missing_values = df['biaya_berobat (Rp)'].isnull().sum()
print("Jumlah data missing pada kolom 'biaya_berobat (Rp)':", missing_values)
```

```python
missing_values = df.isnull().sum()
features_with_missing_data = missing_values[missing_values > 0]
print("Fitur dengan data missing:")
print(features_with_missing_data)
```

```python
# prompt: Ubah klinik bhakti ip kosakata garut menjadi klinik garut

df['rumah'] = df['rumah'].str.replace('klinik bhakti ip kosaka garut', 'klinik bhakti ip')
df['rumah'] = df['rumah'].str.replace('lab. prodia garut', 'laboratorium')
df['tipe'] = df['tipe'].str.replace('spesialis saraf (neurologi)', 'spesialis saraf')
df['tipe'] = df['tipe'].str.replace('spesialis penyakit dalam', 'spesialis')
df['tipe'] = df['tipe'].str.replace('spesialis gigi dan mulut', 'spesialis gigi')
df.head(20)
```

```python
def kategori_biaya(biaya):
  if biaya <= 100000:
    return '<100000'
  elif biaya > 100000 and biaya <= 500000:
    return '250.001-500.000'
  elif biaya > 500000 and biaya <= 750000 :
    return '500.001-750.000'
  elif biaya > 750000 and biaya <= 1000000 :
    return '750.001-1.000.000'
  elif biaya > 1000000 and biaya <= 1250000 :
    return '1.000.001-1.250.000'
  elif biaya > 1250000 and biaya <= 1500000 :
    return '1.250.001-1.500.000'
  else:
    return '> 1.500.000'

df['biaya_berobat (Rp)'] = df['biaya_berobat (Rp)'].apply(kategori_biaya)
df.head()
```
